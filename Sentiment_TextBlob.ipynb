## Kollokationen zwischen Figuren
Um einen besseren Eindruck von der Interaktion zwischen zwei Figuren zu erlangen, wird in diesem Schritt die 25 häufigsten Kollokationen pro Paar angezeigt.

```python
# Kontext tokeniseren
def get_tokens(my_original_files):
    from nltk.tokenize import word_tokenize 
    wordsFiltered = []
    for i in my_original_files:
    
        
        tokens = word_tokenize(i) 
        for w in tokens:
            if w.isalpha():
            
                wordsFiltered.append(w)
                
    return wordsFiltered
```

```python
# Kollokationen herausfiltern und als Liste von Tuple speichern
def get_collocations(tokens):
    import nltk
    tokens = nltk.Text(tkn for tkn in tokens)
    finder = nltk.collocations.BigramCollocationFinder.from_words(tokens)
    bigram_measures = nltk.collocations.BigramAssocMeasures()
    list_of_tuples = finder.score_ngrams(bigram_measures.likelihood_ratio)
    
    return list_of_tuples
```

```python
# aus der zuvor generierten Liste die n häufigsten Kollokationen in ein Dataframe speichern
def get_most_num_collocations(list_of_tuples, num):
    list_of_tuples = list_of_tuples[:num]
    tuples_dict = dict(list_of_tuples)
    df_tuples = pd.DataFrame.from_dict(tuples_dict, orient='index')
    
    return df_tuples
```

```python
# Optional: Um Kollokationen ohne Namen zu erhalten, wird diese Liste von Wörtern aus den Daten kurzzeitig gelöscht
drop_names = ['harry', 'hermine', 'potter', 'granger', 'weasley', 'ron', 'ronald', 'seamus', 'finnigan', 'blaise', 'zabini', 'draco', 'tom', 'riddle', 'ginny', 'voldemort']
```

```python
# Optional: Wörter löschen
df_harry_voldemort['-5'] = df_harry_voldemort['-5'].str.replace(r"(\b" + '\\b|\\b'.join(drop_names) + r"\b)", 'nan', regex=True)
```

```python
# Kontextwörter für Paar in Liste speichern
harry_voldemort = list(df_harry_voldemort["-5"].values)
```

```python
# Tokensieren
tokens_harry_voldemort = get_tokens(harry_voldemort)
```

```python
# fehlerhafte Tokens löschen
drop_name = ['nan', 'j']
```

```python
# tokens von Liste löschen
tokens_harry_voldemort = [ elem for elem in tokens_harry_voldemort if elem not in drop_name ]
```

```python
# Kollokationen herausfiltern
harry_voldemort_tuples = get_collocations(tokens_harry_voldemort)
```

```python
# die 25 häufigsten anzeigen lassen
collocations_harry_voldemort = get_most_num_collocations(harry_voldemort_tuples, 25)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(12,5))
ax = sns.barplot(x=0, y=collocations_harry_voldemort.index, data=collocations_harry_voldemort)
plt.title("Original: Harry Voldemort collocations")
plt.savefig("Sentiment/Original/Kollokationen/harry_voldemort_collocations.png", bbox_inches='tight')
```
