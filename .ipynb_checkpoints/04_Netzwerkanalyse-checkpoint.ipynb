## Ziel der Analyse
Das Ziel soll es sein, die Beziehungen zwischen den einzelnen Charakteren zu visualisieren –  und zwar in der Art und Weise, dass man leicht erkennen kann, welche Figuren im Roman besonders oft zusammen auftauchen und welche eher selten. Dafür schaut man sich die Originaltexte und die Fanfictiontexte an und untersucht für jede Figur A, wie oft eine Figur B in einem Abstand von höchstens X Wörtern zu Figur A genannt wird. Je öfter die Figuren A und B zusammen genannt werden, desto stärker ist ihre Beziehung zueinander. Die Stärke der Beziehung wird in einem Score ausgedrückt, welcher angibt, wie oft die beiden Figuren zusammen erscheinen. Der Graph wird dann vom Score dahingehend beeinflusst, dass Kanten mit höherem Score dicker sind und Knoten mit einer dickeren Kante näher aneinander liegen.

```python
def remove_none_alphabetics(string):
    '''
    Returns list of clean sentences.

            Parameters:
                    string(str): text
    '''
    
    import re
   
    characters = set("?.!/;,:«»–<>_)(][*+&%$´'’")
    sentence = ''.join(c for c in string if  c not in characters)
    
    sentences = (" ".join(re.findall(r"[A-Za-z0-9üäöÜÄÖß]*", sentence))).replace("  "," ")

        
    return sentences
```

Alle Zeichen entfernen, die keine Buchstaben darstellen.

```python
df["network_text"] = df["removed_stop_word_lower"].apply(remove_none_alphabetics)
```

```python
df.to_csv('datasets/df_original.csv')
```

```python
my_files = list(df["network_text"].values)
```

```python
def get_tokens(my_original_files):
    '''
    Returns list of tokens.

            Parameters:
                    my_original_files (list): list of text
    '''
    from nltk.tokenize import word_tokenize 
    wordsFiltered = []
    for i in my_original_files:
    
        
        tokens = word_tokenize(i) 
        for w in tokens:
            if w.isalpha():
            
                wordsFiltered.append(w)
                
    return wordsFiltered
```

### 1.3 Verbindungen von Paaren erstellen <a class="anchor" id="paar_ori"></a>
Für ein funktionierendes Skript braucht man zuerst die Paare von Figuren, deren Beziehungen untersucht werden sollen. Doch wie viele und welche unterschiedlichen Paare gibt es für die Liste von Charakteren überhaupt? Für dieses Problem gibt es das nette Python-Package itertools. Damit kann man innerhalb des Skripts über alle Kombinationen aus Figuren iterieren und dann Scores berechnen.

```python
#sucht alle Paare und deren Verbindungen 
#mit itertools wird über alle Kombinationen aus Figuren iteriert und Scores berechnet

def get_pair_combinations (list_to_iterate, text_files, umkreis):
    '''
    Returns df with Score and Tuples.

            Parameters:
                    list_to_iterate(list): names to search for
                    text_files(list): list of tokens/text
                    umkreis(int): range of words between two names
    '''

    from itertools import combinations
    char=[]
    score=[]
    words = []
    for paar in combinations(list_to_iterate,2):
        zw_summe = 0
        for i in range(len(text_files)):
            if text_files[i]==paar[0]:
                theContext = [x for x in text_files[i-umkreis:i+umkreis]]
                for k in theContext:
                        if k == paar[1]:
                            zw_summe+=1
                            
        char.append(paar)
        score.append(zw_summe)
        
        
    df = pd.DataFrame({"char":char,
                   "score":score})
    return df
```

Erst werden die zwei Listen char und score definiert, in die später jeweils das Charakterpaar und der dazu gehörende Score geschrieben wird. Dann iteriert man über die Liste der Kombinationen von Figuren. Nimmt man als Beispiel die Kombination Harry und Ron. Das Skript durchsucht den Buchtext nach Harry. Für jeden Eintrag namens Harry schaut es in einem Umkreis von 50 Wörtern vor und nach dem Eintrag, wie oft auch Ron genannt wird. Für jeden Treffer erhöht sich die Zwischensumme (zw_summe) um 1. Zum Schluss wird die Liste char um das Tupel (Harry, Ron) erweitert und die Liste score um die dazu gehörige Trefferzahl. Zum Schluss werden beide Listen in den DataFrame df geschrieben.

```python
#Anwenden der Funktion 
df_combi = get_pair_combinations(char_list, hpall_str_split, 50)
```

```python
# Top 10 der häufigsten Kombinationen
df_combi.sort_values(by="score",ascending=False)[:10]
```

### 1.4 Netzwerk erstellen <a class="anchor" id="network_ori"></a>

Berechnete Scores laden

```python
import pandas as pd 
df_combi = pd.read_csv('Network/hp_all_score.csv')
```

```python
# Graph wird definiert und dann mit matplotlib gezeichnet
# Zuerst generiert man einen leeren ungerichteten Graphen G
import networkx as nx
G = nx.Graph()
from ast import literal_eval
#Als nächstes generiert man zwei Listen. Eine mit Kanten und eine mit Gewichtungen. Dabei nimmt man jedoch nur Kombinationen von Figuren, welche mindestens einmal zusammen vorkommen – also einen Score größer 0 besitzen.
edges = list(df_combi.loc[(df_combi.score>0)].char.apply(literal_eval))
weights = list(df_combi.loc[(df_combi.score>0)].score)
#Nun kann man die Kanten dem Graphen G hinzufügen.
G.add_edges_from(edges)
```

```python
G.nodes()
```

```python
nx.draw(G)
```

```python
#Diesen kann man jetzt visualisieren, wobei die Gewichtungen erst hier an das Objekt übergeben werden.
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
g = nx.draw(G,pos=nx.spring_layout(G),
        width=[(weight/2)**0.5 for weight in weights],
        with_labels=True,edge_color="green",
        node_size=[(weight*15000)**0.5 for weight in weights],
        node_color="lightgrey",
        node_shape="o",
        font_size=8,
        font_color="black",
        alpha=0.9)



plt.savefig('Network/hp_all_names.png')
```

Man sieht klar, dass Harry der Protagonist des Buches ist, mit starken Beziehungen zu Ron und Hagrid – aber im Grunde genommen verbunden mit allen anderen Figuren. Andererseits erkennt man, welche Figuren nichts miteinander zu tun haben – wie zum Beispiel Bellatrix und Horace. Zwischen den beiden könnte es aber eine Verbindung geben, wenn man den oben definierten Suchbereich von 50 auf 250 Wörter vergrößern würde. Ab einer gewissen Größe wären alle Figuren miteinander vernetzt, dies würde der Graphenanalyse allerdings ein wenig den Sinn nehmen.
Auffällig ist auch, dass Blaise als Nebenfigur am Rand des Graphen auftaucht und nur fünf dünne Verbindungen zu anderen Figuren besitzt.

### 3.2 Average Clustering Coefficient <a class="anchor" id="average"></a>

Ein anderer Weg, den Unterschied zwischen beiden Datensätze zu evaluieren, ist die Berechnung des Clusterkoeffizienten. Dieser bildet in der Netzwerkanalyse das Maß für die Cliquenbildung bzw. Transitivität in einem Netzwerk. Dabei gilt, wenn alle Nachbarknoten eines Knotens miteinander verbunden sind, dann bilden sie eine Clique. 

```python
print(nx.average_clustering(G_two))
```

```python
print(nx.average_clustering(G))
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))


height = [0.9180586906544489, 0.8900233553595756]
bars = ['Original', 'Fanfiction']
y_pos = np.arange(len(bars))

# Create bars
bar = plt.bar(y_pos, height, color=[ 'darkorange'] )
 
# Create names on the x-axis
plt.xticks(y_pos, bars)

label = ['0.918', '0.890']

# Text on the top of each barplot
def autolabel(rects):
    for idx,rect in enumerate(bar):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.0*height,
                label[idx],
                ha='center', va='bottom')

autolabel(bar)
plt.title("Average Clustering Coefficient")

# Show graphic

plt.savefig('Network/hp_average_clustering.png', bbox_inches='tight')
```

Bei gleicher Anzahl an Charakteren zeigen die Originaltexte einen höheren Clusteringkoeffizienten, also besitzen sie demnach eine höhere Wahrscheinlichkeit für Cliquenbildung der Figuren. Trotzalledem ist die Differenz der beiden Zahlen nicht sehr gravierend.

### 3.4 Closeness Centrality <a class="anchor" id="close"></a>

Die Closeness Centrality ist ein Maß für die Zentralität in einem Netzwerk. Dieses wird durch den Kehrwert der Summe aller kürzesten Pfade zwischen einem Knoten zu anderen Knoten berechnet. Dabei gilt, je zentraler ein Knoten ist, desto näher ist er allen anderen.

```python
import collections
from collections import OrderedDict

c=nx.closeness_centrality(G)

keys = c.keys()
values = c.values()
fig = plt.figure(figsize=(5,5))
plt.xticks(range(len(c)), list(keys), rotation = 90)
bar = plt.bar(keys, values)

plt.title("Closeness Centrality Original")

# Show graphic

plt.savefig('Network/hp_original_centrality.png', bbox_inches='tight')
```

Auch hier besitzen Harry, Ron, Hermine und Snape die höchste Zentralität.

```python
import collections
from collections import OrderedDict

c=nx.closeness_centrality(G_two)

keys = c.keys()
values = c.values()
fig = plt.figure(figsize=(5,5))
plt.xticks(range(len(c)), list(keys), rotation = 90)
bar = plt.bar(keys, values)

plt.title("Closeness Centrality Fanfiction")

# Show graphic
#plt.show()
plt.savefig('Network/hp_fanfiction_centrality.png', bbox_inches='tight')
```

Die Zentralität zeigt, wie auch schon das obige Netzwerk ergeben hat, dass Horace am weitesten entfernt von all anderen Knoten ist und Sirius eine zentrale Figur in den Fanfictiontexten darstellt.
