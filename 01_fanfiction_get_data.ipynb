## 1 Fanfictiondaten generieren <a class="anchor" id="dataset"></a>
### 1.1 Texte einlesen <a class="anchor" id="text"></a>

```python
import glob
import shutil
from pathlib import Path
```

```python
# alle Fan-Fiction Texte werden eingelesen
p = Path('./Fan-Fiction/TEI')
files = list(p.glob('**/*'))
```

### 1.2 Harry Potter herausfiltern <a class="anchor" id="hp"></a>

```python
# alle Texte, die Harry Potter beinhalten in Liste speichern
list_of_strings_to_search_for = ['Harry Potter']
counter = 0
my_files = []
for file in files:
    with open(file, encoding="utf-8") as fin:
        text = fin.read()
        for s in list_of_strings_to_search_for:
            if s in text:
                print("%s is matched in %s" % (s, file))
                counter += 1
                my_files.append(file)
                if counter >= 200:
                    break
```

```python
len(my_files)
```

### Daten abspeichern <a class="anchor" id="save"></a>

```python
# 9 811 Texte aus Liste kopieren und in neuen Ordner speichern
dest_dir = Path('./Fanfictions')
new_files = []
for filename in my_files:
    new_folder = shutil.copy(filename, dest_dir)
    new_files.append(new_folder)
```
